```python
from html.parser import HTMLParser

class HTMLFilter(HTMLParser):
    text = ""
    def handle_data(self, data):
        self.text += data
```

```python
paths = ['Balance Sheets', 'Cash Flow', 'Income Statement', 'Notes', 'Others']
arr = []

for path in paths:
  files = glob(f'/content/TableClassifierQuaterlyWithNotes/{path}/*?.html')

  for fil in files:
    htmlData = HTMLFilter()

    with open(fil, 'r') as doc:
      text = doc.read()

    htmlData.feed(text)
    txt = htmlData.text
    arr.append({'Text': txt, 'Label': 'Financial_Statements'})

  print(f'Read all {path} files')
```

```python
finance_df = pd.DataFrame(arr)
finance_df.to_csv('/content/dataset/finance_dataset.csv')
```

```python
folders = glob('/content/topics/*')
arr = []

for folder in folders:
  files = glob(f'{folder}/*.txt')

  count = 0
  for fil in files:
    with open(fil, 'r') as doc:
      text = doc.read()

    arr.append({'Text': text, 'Label': 'Creative_Writing'})
    count += 1

    if count == 35:
      break

  print(f'{len(arr)} files completed')
```

```python
poems_files_dataset = pd.DataFrame(arr)
poems_files_dataset.to_csv('/content/dataset/poems_files_dataset.csv')
```

# Combining the CSV files

```python
files = glob('/content/dataset/*?.csv')
```

```python
files
```

```python
for f in range(len(files)):
  if f == 0:
    df = pd.read_csv(files[f], index_col=0)
  else:
    df = pd.concat([df, pd.read_csv(files[f], index_col=0)], axis=0)

  print(f'Read {files[f]} dataset with shape of {df.shape}')
```

```python
df
```

```python
df['Label'] = df['Label'].replace({
    'Politics': 'Government',
    'Technologie': 'Technology',
    'Tech': 'Technology',
    'Entertainment': 'Creative_Writing',
    })
```
